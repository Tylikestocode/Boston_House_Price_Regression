## Feature Engineering

```python
data_2 = data.copy()
data_3 = data.copy()
data_2 = data_2.drop(['MEDV'], axis=1)
```

```python
valid_features = []
for feature in data_2.columns:
    if np.min(data_2[feature]) != 0:
        valid_features.append(feature)
```

```python
data_2_log = np.log(data_2[valid_features])
data_2_sqrt = np.sqrt(data_3[valid_features])
```

```python
for feature in data_2_log.columns:
    old_corr = np.corrcoef(data[feature], data['MEDV'])[0][1]
    log_corr = np.corrcoef(data_2_log[feature], data['MEDV'])[0][1]
    print(f'{feature} Old Correlation : {old_corr:.2f}')
    print(f'{feature} Log Correlation : {log_corr:.2f}')
    print('----------------------------------')
```

```python
for feature in data_2_sqrt.columns:
    old_corr = np.corrcoef(data[feature], data['MEDV'])[0][1]
    log_corr = np.corrcoef(data_2_sqrt[feature], data['MEDV'])[0][1]
    print(f'{feature} Old Correlation : {old_corr:.2f}')
    print(f'{feature} Sqrt Correlation : {log_corr:.2f}')
    print('---------------------------------')
```

## Data Preprocessing

```python
data['LSTAT'] = np.log(data['LSTAT'])
data['CRIM'] = np.sqrt(data['CRIM'])
```

## Splitting the Data

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(['MEDV'],axis=1)
y = data['MEDV']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)

print(f"X_train Shape : {X_train.shape}")
print(f"X_test Shape : {X_test.shape}")
print(f"y_train Shape : {y_train.shape}")
print(f"y_test Shape : {y_test.shape}")
```

## Feature Extraction

```python
def pearson_elimination(x, num_of_features):
    correlations = []
    features = X.columns
    for feature in  X.columns:
        correlation = np.corrcoef(X[feature], y)[0][1]
        correlations.append(correlation)
    Ranked_Features = [features[i] for i in np.argsort(correlations)]
    print(Ranked_Features[:num_of_features])
    return Ranked_Features[:num_of_features]
```

```python
Features = pearson_elimination(X_train, num_of_features=3)
```

## Model Creation

```python
from sklearn.linear_model import LinearRegression
```

```python
X_train = X_train[Features]
X_test = X_test[Features]
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
print(f'Coefficients : ', linreg.coef_)
print(f'Intercept : ', linreg.intercept_)
```

### Linear Model : $\hat{y} = 65.633443598404 -11.78015081 \theta_{1} -0.86385106 \theta_{2} + 0.0649303 \theta_{3}$

## Model Evaluation

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
y_pred = linreg.predict(X_test)

r2 = r2_score(y_true=y_test, y_pred=y_pred)
mse = mean_squared_error(y_true=y_test, y_pred=y_pred)
print(f'Model R2 Score : {r2:.2f}')
print(f'Model MSE : {mse:.2f}')
```
